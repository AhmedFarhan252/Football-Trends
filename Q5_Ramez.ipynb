```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

#Loading data
df = pd.read_pickle('Major_International_Results.pkl')
wc = df[df['tournament'] == 'FIFA World Cup']
df.head()
```

# => Made Columns for Home Wins, Away Wins and Draws
- Neutral columns was made to be equal to False so that

```python
# Made a new DataFrame in which neutral = False
# Going to Use this to add new Columns and do EDA

newDF = df[df['neutral'] == False]
```

## => What are the trends in goals scored at home vs goals scored away?

```python
dfdf = newDF.loc[: , 'home_team':'away_score']

plt.figure(figsize=(15,5))
sns.barplot(x=dfdf.home_team[:300],y=dfdf.home_score[:300],data=dfdf,color='green',label='home_scores')
sns.barplot(x=dfdf.home_team[:300],y=dfdf.away_score[:300],data=dfdf,color='red',label='away_scores')
plt.xlabel('TEAMS',fontsize = 15,color='black')
plt.xticks(rotation=90)
plt.legend()
plt.ylabel('SCORES',fontsize = 15,color='black')
plt.title('AWAY SCORES VS HOME SCORES IN HOME GAMES',fontsize = 15,color='black')
plt.grid()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=dfdf.away_team[:300],y=dfdf.home_score[:300],data=dfdf,color='green',label='home_scores')
sns.barplot(x=dfdf.away_team[:300],y=dfdf.away_score[:300],data=dfdf,color='red',label='away_scores')
plt.xlabel('TEAMS',fontsize = 15,color='black')
plt.xticks(rotation=90)
plt.legend()
plt.ylabel('SCORES',fontsize = 15,color='black')
plt.title('AWAY SCORES VS HOME SCORES IN AWAY GAMES',fontsize = 15,color='black')
plt.grid()
```
